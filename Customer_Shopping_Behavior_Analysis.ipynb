```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
if __name__ == '__main__':
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')
    df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# create a column age_group

labels =['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create  column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Daily':90,
    'Occasional':14,
    'Annually':365,
    'Every 3 Month':90
}
df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
# Define a mapping dictionary
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Daily':90,
    'Occasional':14,
    'Annually':365,
    'Every 3 Month':90
}

# Fill missing values using the mapping
df['purchase_frequency_days'] = df['purchase_frequency_days'].fillna(
    df['frequency_of_purchases'].map(frequency_mapping)
)
```

```python
df = df.astype(str)
df = df.fillna('discount_promo_code')

```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus   # password मध्ये '@' असल्याने encoding आवश्यक

# MySQL details
username = "root"
password = quote_plus("root@10")    # @ मुळे error येतो म्हणून encode केलं
host = "localhost"
port = 3306
database = "customer_behavior"

# ✅ EXACT तुझ्या style मध्ये — MySQL साठी connection string
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# टेबल नाव
table_name = "customer_behavior"

# DataFrame → MySQL
df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f"Data Successfully inserted into table {table_name} in database {database}")
```
